## Sucesión de Fibonacci

```python
#Sucesión_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

## Optimización Sucesión de Fibonacci

```python
# Defino la función optimizada
def fib_optimizado(n):
    if n < 2:
        return 1
    a, b = 1, 1   # F(0)=1, F(1)=1
    # Lo resuelvo por intercambio de variables repetitivamente 
    for _ in range(2, n+1):
        a, b = b, a + b
    return b

#Calculo el resultado y el tiempo de ejecucion para el algoritmo recursivo
start = time.time()
print(Fibonacci(40))
end = time.time()
print("Algoritmo Recursivo - Tiempo de ejecución:", end - start, "segundos")

#Calculo el resultado y el tiempo de ejecucion para el algoritmo optimizado
start = time.time()
print("")
print(fib_optimizado(40))
end = time.time()
print("Algoritmo Optimizado - Tiempo de ejecución:", end - start, "segundos")
```

Podemos ver que esta versión optimizada de la sucesión de Fibonacci es mucho mejor que el algoritmo recursivo original. Mientras que la versión recursiva calcula múltiples veces los mismos términos, provocando una complejidad exponencial, la versión iterativa calcula cada término una única vez utilizando un bucle. De esta manera, la complejidad se reduce a O(N). El resultado es un algoritmo mucho más eficiente, especialmente cuando usamos valores grandes de N.

## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

# Calculamos los tiempos de ejecucion del algoritmo
start = time.time()
print("")
reinas(8)
end = time.time()
print("Algoritmo - Tiempo de ejecución:", end - start, "segundos")
print("")
```

## Optimización N-Reinas

```python
def reinas_opt(N):
    
    # Guarda la fila (1..N) donde se coloca la reina en cada columna
    solucion = [0] * N

    # Creamos estructuras auxiliares para verificar las restricciones
    
    # -------------------------------
    # OPTIMIZACIÓN 1 - Filas en sets
    # -------------------------------
    # En vez de comprobar si hay 2+ reinas en la misma fila utilizando un bucle con un count, 
    # guardamos las filas ocupadas en un set
    cols = set()

    # ------------------------------------
    # OPTIMIZACIÓN 2 - Diagonales en sets
    # ------------------------------------
    # En vez de comprobar si hay 2+ reinas en la misma diagonal utilizando dos bucles, 
    # guardamos las diagonales ocupadas en sets 
    
    # Diagonales principales (columna - fila es constante) 
    diag1 = set()
    # Diagonales secundarias (columna + fila es constante)
    diag2 = set()

    def bt(etapa):  
        # Si etapa == N, se han colocado N reinas correctamente (una por columna) 
        if etapa == N:
            print(solucion)   # Se imprime la solución válida en pantalla
            print()
            return
            
        # En esta columna 'etapa', se intenta colocar la reina en todas las filas posibles (1..N)
        # Generamos el árbol de búsqueda
        for fila in range(1, N + 1):

            # Calculamos los "ID" de las diagonales para esta posición
            # Si otra reina tiene el mismo ID, comparten diagonal
            d1 = etapa - fila     # diagonal principal
            d2 = etapa + fila     # diagonal secundaria

            # ----------------------------------------------------
            # OPTIMIZACIÓN 3 - PODA
            # ----------------------------------------------------
            # Antes, el algoritmo original recorría y comprababa las posibles soluciones ya generadas (costoso)
            
            # Ahora, con los sets, vamos a hacer 3 preguntas:
            # - ¿la fila ya está ocupada?
            # - ¿la diagonal principal ya está ocupada?
            # - ¿la diagonal secundaria ya está ocupada?
            # Si la respuesta es sí a cualquiera, entonces descartamos esta opción inmediatamente 

            # Evitamos los bucles del algoritmo original (y evitamos crear la función es_prometedora)
            if fila in cols or d1 in diag1 or d2 in diag2:   
                continue

            # De no ser asi, guardamos que en la columna 'etapa' pusimos una reina en 'fila'
            # y marcamos fila/diagonales como ocupadas
            solucion[etapa] = fila
            cols.add(fila)
            diag1.add(d1)
            diag2.add(d2)

            # Y pasamos a la siguiente columna para colocar la siguiente reina llamando recursivamente al algoritmo
            bt(etapa + 1)

            # BACKTRACKING
            # Ya se ecploraron todas las posibilidades (o esa rama no llevo a solucion)
            # Quitamos la reina colocada, liberamos fila y diagonales
            cols.remove(fila)
            diag1.remove(d1)
            diag2.remove(d2)
            solucion[etapa] = 0

    # Llamada inicial al backtracking
    # Comienza en la primera columna
    bt(0)


# Calculamos los tiempos de ejecucion del algoritmo optimizado
start = time.time()
print("")
reinas_opt(8)
end = time.time()
print("Algoritmo Optimizado - Tiempo de ejecución:", end - start, "segundos")
print("")
```

El algoritmo original de N-Reinas utiliza backtracking, pero en cada paso verifica si una solución parcial es prometedora recorriendo repetidamente la lista de posiciones ya colocadas. En particular, emplea bucles para detectar filas repetidas y bucles anidados para comprobar diagonales, lo que incrementa el coste de verificación. 

En la versión optimizada se mantiene el mismo enfoque de backtracking, pero se incorporan estructuras auxiliares (con los sets) para almacenar filas y diagonales ya ocupadas. Gracias a esta modificacion, la comprobación de restricciones se realiza en tiempo constante O(1), ya que basta con consultar si la fila o diagonal ya está registrada en los conjuntos. Como consecuencia, aunque la complejidad del problema sigue siendo exponencial (comun en backtracking), el tiempo total de ejecución disminuye significativamente en la práctica, debido a la reducción significativa del coste por cada estado explorado y a una poda más eficiente del árbol de búsqueda.

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
